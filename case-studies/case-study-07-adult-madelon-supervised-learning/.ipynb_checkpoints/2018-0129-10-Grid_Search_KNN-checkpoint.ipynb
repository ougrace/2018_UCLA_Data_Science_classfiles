# Observations

###  Uniform weights for neighbors is strongly affected by odd / even.
For some reason we are getting this peak/valley effect with the Uniform weighted scheme (green and blue lines).  Uniform literally performs better on odd numbered neighbors than even.  I can't really fathom why this would be the case.  Could be an implementation bug?  We see the effect across both datasets

### Madelon
In the madelon data we see that both "weighted" schemes for neighbors outperforms the uniform scheme in general.  The L1 and L2 distances do not seem to make much difference. Question - what does this say about the linearity of the dataset?

The number of neighbors has a sweet spot around 15-20

### Adult
In the adult dataset we see that the L1 distance metric outperforms the L2 distance.  However the choice between "uniform" and "weighted" does not matter much.
The sweet spot for neighbors seems to be around 20 as well.
