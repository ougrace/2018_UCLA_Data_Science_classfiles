### Adult

```python
def plot_xplained_var(dataset, pca_obj):
    plt.figure(figsize = (12,4))
    plt.plot(np.cumsum(pca_obj.explained_variance_ratio_))
    plt.title("Explained Variance Ratio for PCA Components: {} Dataset".format(dataset))
    plt.xlabel("Number of PCA Components")
    plt.ylabel("Explained Variance Ratio")
```

```python
def plot_scree(dataset, pca_obj):
    plt.figure(figsize = (12,4))
    plt.plot(pca_obj.explained_variance_ratio_)
    plt.title("Explained Variance Ratio for PCA Components: {} Dataset".format(dataset))
    plt.xlabel("Number of PCA Components")
    plt.ylabel("Explained Variance Ratio")
```

```python
def plot_kurtosis_ica(dataset_name, data, n = None):
    
    ica = FastICA(n, max_iter=600)
    S = ica.fit_transform(data)
    
    ic_comp = pd.DataFrame(S)
    ic_comp = ic_comp.kurt(axis=0).abs()
    ic_comp.sort_values(ascending = False, inplace = True)
    
    
    plt.figure(figsize = (16,4))
    plt.bar(range(len(ic_comp)), ic_comp)
    plt.title("Kurtosis by ICA components: {} Dataset".format(dataset_name))
    plt.xlabel("ICA Components")
    plt.ylabel("Kurtosis, 0 == Gaussian")
```

```python
plot_kurtosis_ica("Madelon", madelon_train_f)
```

```python
plot_kurtosis_ica("Madelon", madelon_train_f, 100)
```

```python
plot_kurtosis_ica("Adult Numeric Only", adult_train_no_dummy[cont_columns])
```
