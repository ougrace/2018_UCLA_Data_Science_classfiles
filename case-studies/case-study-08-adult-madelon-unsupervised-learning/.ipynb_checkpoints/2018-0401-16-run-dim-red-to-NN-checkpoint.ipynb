```python
from tqdm import tqdm
from lib.bic import bic

def fit_and_time(model, data, variance='kmeans'):
    start = time()
    model = model.fit(data)
    end = time() - start
    return {'fit_time_'+variance : end, 'model' : model}

def process_results(results_list, data, variance='kmeans'):
    df = pd.DataFrame(results_list)
    if type(results_list[0]['model']) == KMeans:
        df['k'] = df.model.apply(lambda x: x.n_clusters)
        df['bic'] = df.model.apply(lambda x: bic(x, data))
    else:
        df['k'] = df.model.apply(lambda x: x.n_components)
        df['bic'+variance] = df.model.apply(lambda x: x.bic(data))
    df.set_index('k', inplace=True)
    return df 
```

```python
def fit_clusters(ks, train_data):
    kmeans_models_adult, \
    gm_full_models_adult, \
    gm_tied_models_adult, \
    gm_diag_models_adult, \
    gm_spherical_models_adult = [], [], [], [], []
    for k in tqdm(ks):
        kmeans_models_adult.append(fit_and_time(KMeans(n_clusters=k, init="k-means++"), train_data))
        gm_full_models_adult.append(fit_and_time(GaussianMixture(n_components=k, covariance_type='full', random_state=0),
                                                 train_data, 'full'))
        gm_tied_models_adult.append(fit_and_time(GaussianMixture(n_components=k, covariance_type='tied', random_state=0),
                                                 train_data, 'tied'))
        gm_diag_models_adult.append(fit_and_time(GaussianMixture(n_components=k, covariance_type='diag', random_state=0, reg_covar=1E-5),
                                                 train_data, 'diag'))
        gm_spherical_models_adult.append(fit_and_time(GaussianMixture(n_components=k, covariance_type='spherical', random_state=0),
                                                      train_data, 'spherical'))

    kmeans_models_df = process_results(kmeans_models_adult, train_data)
    gm_full_models_df = process_results(gm_full_models_adult, train_data, 'full')
    gm_tied_models_df = process_results(gm_tied_models_adult, train_data, 'tied')
    gm_diag_models_df = process_results(gm_diag_models_adult, train_data, 'diag')
    gm_spherical_models_df = process_results(gm_spherical_models_adult, train_data, 'spherical')
    gm_models_df = pd.merge(kmeans_models_df, gm_full_models_df,left_index=True, right_index=True)
    gm_models_df = pd.merge(gm_models_df, gm_tied_models_df,left_index=True, right_index=True)
    gm_models_df = pd.merge(gm_models_df, gm_diag_models_df,left_index=True, right_index=True)
    gm_models_df = pd.merge(gm_models_df, gm_spherical_models_df,left_index=True, right_index=True)
    return gm_models_df
```
