```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier

from lib.benchmark import run_model
from tqdm import tqdm
```

# Adult 

```python
test_results = {}
percentages = [1,2,3,4,5,7,10,15,20,25,30,40,50,60,70,80,90,100]

tests = [('logistic regression', 'basic', LogisticRegression()),
         ('logistic regression', 'scaled_numeric', LogisticRegression()),
         ('logistic regression', 'skew_normalized_numeric', LogisticRegression()),
         ('logistic regression', 'scaled_all', LogisticRegression()),
         ('logistic regression', 'skew_normalized_all', LogisticRegression()),
         ('knn', 'basic', KNeighborsClassifier()),
         ('knn', 'scaled_numeric', KNeighborsClassifier()),
         ('knn', 'skew_normalized_numeric', KNeighborsClassifier()),
         ('knn', 'scaled_all', KNeighborsClassifier()),
         ('knn', 'skew_normalized_all', KNeighborsClassifier()),
         ('decision tree', 'basic', DecisionTreeClassifier()),
         ('decision tree', 'scaled_numeric', DecisionTreeClassifier()),
         ('decision tree', 'skew_normalized_numeric', DecisionTreeClassifier()),
         ('decision tree', 'scaled_all', DecisionTreeClassifier()),
         ('decision tree', 'skew_normalized_all', DecisionTreeClassifier())]
```

```python
for t in tests:
    for n in tqdm(percentages):
        test_name = "{} {} {}".format(t[0], t[1], n)
        data_set = t[1]
        model = t[2]
        test_results[test_name] = run_model(model, test_name, n,
                                                data['adult']['train'][data_set],
                                                data['adult']['train']['labels'])
```

```python
test_results = pd.DataFrame(test_results).T.sort_values('n')
```

```python
fig = plt.figure(figsize=(20,4))
models = ['logistic regression', 'knn', 'decision tree']

for i, m in enumerate(models):
    fig.add_subplot(1,3,1+i)

    for t in [(t[0]+' '+t[1]) for t in tests if m in t[0]]:

        test = test_results.loc[test_results.index.str.contains(t)]
        plt.plot(test.n_pcnt, test.f1_test_score, label=t)
    plt.ylim(0.5,0.7)
    plt.legend()
    plt.title(m)
```
